```python
def plot_df(df, x1,x2,x3, y1,y2,y3,title="", xlabel='Date', ylabel='Revenue', dpi=500):
    plt.plot(x1, y1,"-b",label="US")
    plt.plot(x2, y2,"-r",label="UK")
    plt.plot(x3, y3,"-g",label="SW")
    plt.legend()
    plt.ylim(-1.5, 200)    
    plt.show()
```

```python
plot_df(df,df1.dt,df2.dt,df3.dt,df1.total_revenue,df2.total_revenue,df3.total_revenue)
```

```python
def plot_df(df, x1,x2,x3, y1,y2,y3,title="", xlabel='Date', ylabel='Revenue', dpi=500):
    plt.plot(x1, y1,"-b",label="United States")
    plt.plot(x2, y2,"-r",label="United Kingdom")
    plt.plot(x3, y3,"-g",label="Switzerland")
    plt.legend()
    plt.xlabel("Date")
    plt.ylabel("Revenue")
    plt.title("Daily Revenue Trend of three countries from 15th February 2027 to 21st February 2007")
    plt.ylim(-1.5, 200)    
    plt.xticks(rotation=90, fontweight='light',  fontsize='x-small',)
    plt.show()
```

```python
plot_df(df,df1.dt,df2.dt,df3.dt,df1.total_revenue,df2.total_revenue,df3.total_revenue)
```
